```python
from sklearn.datasets import load_iris 
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
iris = load_iris() 
X = iris.data[:, 2:] # petal length and width y = iris.target
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X,y)

```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (2,2), dpi=500)
tree.plot_tree(tree_clf,
              feature_names = fn, 
               class_names=cn,
               filled = True)
fig.savefig('decision_tree_chapt6.png')
```

```python
tree_clf.predict_proba([[5,1.5]])# predice la prob de que pertenezca a cualquiera de las instancias 
```

```python
tree_clf.predict([[5,1.5]])
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2, criterion='entropy')
tree_clf.fit(X,y)

fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (2,2), dpi=500)
tree.plot_tree(tree_clf,
              feature_names = fn, 
               class_names=cn,
               filled = True)
fig.savefig('decision_tree_chapt6_entropy.png')
```

```python
# decision regression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
tree_reg = DecisionTreeRegressor(max_depth=2)
m = 100 
X = 6 * np.random.rand(m, 1) - 3 
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

tree_reg.fit(X, y)
np.mean(y)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (2,2), dpi=500)
tree.plot_tree(tree_reg,
               filled = True)
fig.savefig('regression_tree_chapt6.png')# value es el valor medio de la categora en cuestion
```
