```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression 

mnist = fetch_openml('mnist_784', version=1)
X, y = mnist['data'], mnist['target']
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
# using clustering for semisupervised learning
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])

log_reg.score(X_test, y_test)
```

```python
# mejorar la performance a patir de hace kmeans
import numpy as np
from sklearn.cluster import KMeans


k=50
kmeans = KMeans(n_clusters=k)
X_digits_dist = kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train[representative_digit_idx]
```

```python
y_representative_digits = np.array([4, 8, 0, 6, 8, 3, 7,7,9,2,5,5,8,5,2,1,2,9,6,1\
                                    ,1,6,9,0,8,3,0,7,4,1,6,5,2,4,1,8,6,3,9,2,4,2,9,\
                                    4,7, 6, 2, 3, 1, 1])
```

```python
log_reg1 = LogisticRegression()
log_reg1.fit(X_representative_digits, y_representative_digits)
log_reg1.score(X_test, y_test)
```

```python
# label propagation
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_ == i] = y_representative_digits[i]

```

```python
#DBSCAN- density clustering 
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=5) 
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_
```

```python
dbscan.components_
```

```python
# DBSCAN - utilizarlo con knclassifier- usando los core classifiers
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50) 
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])

knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
```

```python
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2] = -1
y_pred.ravel()
```
